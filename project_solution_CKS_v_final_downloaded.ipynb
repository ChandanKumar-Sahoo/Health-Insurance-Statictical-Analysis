## <span style="font-family: Arial; font-weight:bold;font-size:1 em;color:Orange;"> 4.a. Do charges of people who smoke differ significantly from the people who don't
<a id='SC'></a>

### Step 1: Developing a clear research problem 
* Do charges for smokers differ significantly than non smoker

* This question deals with a categorical attribute and a continous or numerical attribute, thus we can carry out variance test followed by a chi square test to confirm the hypothesis

```python
df.head()
```

```python
# here i have taken age group as 27 for my sample size
tseven = df[df['age'] == 27]
tseven.smoker.value_counts()
```

* Thus there are 19 non smokers and 09 smokers present in 27 years age group, now comparing the variance between them 

```python
sample_non_smoker = tseven[tseven['smoker'] == 'no'].charges.iloc[:-10]   #excluding the last 10 elements to match the size 2 samples

sample_smoker = tseven[tseven['smoker'] == 'yes'].charges
# print(sample_smoker)
# print(sample_non_smoker)
```

```python
df.var() # to cross c
```

```python
# now checking for the variance for smoker and non smoker
vns, vs = np.var(sample_non_smoker), np.var(sample_smoker)
print(vns, vs)
```

* Variance of non smoker is less than smoker, that is smokers has more variability for 27 years of age group people. 
now, we have to check is the difference in variance shown above is statistically significant or not

### Step 2: Stating the null and alternative hypothesis:

* $H_0$: variation in charges of smoker and non_smoker is equal, vs = vns
* $H_A$: variation in charges of smoker is more than non_smoker, vs > vns

### Step 3: Deciding the significance level, sample size and degree of freedom

* Here we considered level of significance, α = 0.05
* sample size, n = 9,
* degree of freedom, dof = n-1 = 8

### Step 4: Identify the test statistic
we shall use chi square test here due to presence of variance
* and the chi_critical value = 15.51  **critical chi_squared statisitics from the chi_squared  table


### Step 5: Calculate chi value

```python
n = 9
dof = n - 1
chi_critical = 15.51
chi = (dof * vns) / vs
print(chi)
if chi < chi_critical:
    print("Since the test statistic is less than the critical value, we fail to reject the null")
else:
    print("Since the test statistic is more than the critical value, we reject the null")
```

### Step 6:  Decide to reject or accept null hypothesis

* In this question, chi value is less than chi squared critical value
* Thus, the statistical decision is to retain the null hypothesis

#### There are sufficiently no evidence to reject the null hypothesis that  variation in charges of smokers and non smokers is equal

```python
sns.boxplot(x = "children", y = "bmi", hue ='sex', data = df)
plt.title('no. of children vs. BMI')
plt.show()
```

* The box plot shows almost dissimilar shapes and spreads with outliers specially for females with 0 child, 1 child, and 2 children

### Step 2: Stating the null and alternative hypothesis:

* $H_0$: $\mu1$ = $\mu2$ = $\mu3$
* $H_A$: At least one $\mu$ differs 

Here $\mu1$, $\mu2$ and $\mu3$ are the mean bmi for women having zero child, one child, and two children respectively.

### Step 3: Decide the significance level

Here we considered level of significance, α = 0.05

### Step 4: Identify the test statistic
Here we have three groups. Analysis of variance can determine whether the means of three or more groups are different. ANOVA uses F-tests to statistically test the equality of means.

### Step 5: Calculate p value using ANOVA table

```python
mod = ols('bmi ~ children', data = df).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print(aov_table)
```

### Step 6:  Decide to reject or accept null hypothesis

* In this question, p value is 0.6412 which is more than our chosen level of significance at 5%
Thus, the statistical decision is to retain the null hypothesis

#### There are sufficiently no evidence to reject the null hypothesis that bmi of women having zero child, one child and two children are different.
